#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_month = fitbit.groupby(['Months_encoded'], as_index=False).Steps.median()
# print(f'Median steps by month = {median_month}')
median_month
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_work = fitbit.groupby(['Work_or_Weekend'], as_index=False).Steps.median()
# print(f'Median steps by work or weekend = {median_work}')

fig, ax = plt.subplots(1,2,figsize=(12,4))
plt.sca(ax[0])
plt.axis([0, 12, 7500, 14000])
plt.bar(median_month.Months_encoded, median_month.Steps, label='median month')
plt.title('Month steps median')
plt.xlabel('Month')
plt.ylabel('Steps median')
plt.xticks(range(1,14),['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec',''])


plt.sca(ax[1])
plt.bar(median_work.Work_or_Weekend, median_work.Steps, label='median work/weekend')
plt.axis([-.5,1.5, 9500, 10600])
plt.title('Work or Weekend steps median')
plt.xlabel('Work or Weekend')
plt.ylabel('Steps median')
plt.xticks(range(0,2),['work','weekend'])



plt.show()
plt.clf()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

def graphique(dataset):
    names = ['Minutes Lightly Active','Minutes Very Active', 'Minutes Sedentary']
    n = len(names)
    plt.figure(figsize=(12,12))
    
    for k, i in zip(names, range(1, n+1)):
        plt.subplot(n,1,i)
        plt.scatter(dataset[k], dataset.Steps)
        
graphique(fitbit)

# fitbit['Steps']
```
